```python
data = [('2022-03-15','2022-03-15 12:34:56.789'),
        ('2022-03-01','2022-03-01 00:00:00.000'),
        ('2022-03-10','2022-03-10 12:34:56.789')]
df = spark.createDataFrame(data, ['date_col', 'timestamp_col'])
df.show()
```

```python
from pyspark.sql.functions import *
df.select('date_col',date_add('date_col',30).alias('add_date')).show()
```

```python
from pyspark.sql.functions import *
df1 = df.select('date_col',date_sub('date_col',10).alias('sub_date')).show()
```

```python
from pyspark.sql.functions import *
df1 = df.select('date_col',date_sub('date_col',10).alias('sub_date'))
df1.select('date_col','sub_date',datediff('date_col','sub_date').alias('diff_date')).show()
```

```python
from pyspark.sql.functions import *
df.select('date_col',year('date_col').alias('year_dd')).show()
```

```python
from pyspark.sql.functions import *
df.select('date_col',month('date_col').alias('month_dd')).show()
```
