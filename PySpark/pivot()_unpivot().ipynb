```python
import pyspark 
from pyspark.sql import SparkSession

data = [('Banana',1000,'usa'),('carrot',1500,'usa'),('Beans',1600,'usa'),\
        ('Orange',2000,'usa'),('Orange',1000,'usa'),('Banana',400,'china'),\
        ('carrot',1200,'china'),('Beans',1500,'china'),('Orange',1000,'china'),\
        ('Banana',2000,'canada'),('carrot',1000,'canada'),('Beans',1000,'mexico')]
        

columns =['product','amount','country']

df_fruits = spark.createDataFrame(data,columns)
df_fruits.show()
```

```python
df_fruits1 = df_fruits.groupBy('product').pivot('country').sum('amount')#.show()
df_fruits1.na.fill(0).show()
```

```python
from pyspark.sql.functions import expr
df_fruits1.select('product',expr("stack(4,'canada',canada,'china',china,'mexico',mexico,'usa',usa) as (country,total)")).show()
```
