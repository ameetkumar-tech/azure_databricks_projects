```python
data = [('2022-03-15','2022-03-16 12:34:56.789'),
        ('2022-03-16','2022-03-16 12:34:56.789'),
        ('2022-03-17','2022-03-16 12:34:56.789')]
df = spark.createDataFrame(data, ['date_col','timestamp_col'])
df.show()
```

```python
from pyspark.sql.functions import *

df.select('date_col',date_format('date_col','yyyy/MM/dd').alias('converted_date')).show()
```
